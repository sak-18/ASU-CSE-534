# Step 1: Initial Setup

## Step 1.1: Importing FABlib library.

```python
from fabrictestbed_extensions.fablib.fablib import FablibManager as fablib_manager
fablib = fablib_manager()
```

## Step 1.2: Fetch the Slice

Get the slice already setup previously for experimenting.

```python
slice = fablib.get_slice(name="p4_route_tracing_v1")
```

# Step 2: Installing the required packages
In this step, we will install the required packages to run the labs. Specifically, we will install the BMv2 software switch and its control plane, the P4 compiler (p4c), and net-tools.


List all the names of the servers and switches.

```python
switch_names = ['switch1', 'switch2', 'switch3', 'switch4', 'switch5', 'switch6']
server_names = ['server1', 'server2', 'server3', 'server4', 'server5', 'server6']
node_names = switch_names + server_names
```

## Step 2.1 Installing BMv2
The BMv2 software switch will be installed on the switch nodes. We will upload the script [scripts/install_bmv2.sh](./scripts/install_bmv2.sh) to the switch and execute it

```python
for switch_name in switch_names:
    print(switch_name)
    switch = slice.get_node(name=switch_name)
    switch.upload_file('scripts/install_bmv2.sh', 'install_bmv2.sh')
    stdout, stderr = switch.execute(f'chmod +x install_bmv2.sh &&  ./install_bmv2.sh',quiet=True)
```

## Step 2.2 Installing net-tools
The net-tools package will be installed on the nodes. This package will allow us to use the ifconfig and the arp commands 

```python
for node_name in node_names:
    print(node_name)
    node = slice.get_node(name=node_name)
    stdout, stderr = node.execute(f'sudo apt-get install -y net-tools', quiet=True)  
```

## Step 2.3 Installing scapy
Installing scapy to be able to craft and send raw packets on the servers


```python
for node_name in node_names:
    print(node_name)
    node = slice.get_node(name=node_name)
    stdout, stderr = node.execute(f'sudo apt-get update && sudo apt-get install -y python3-scapy', quiet=True)  
```

## Step 2.4 Installing traceroute
Installing traceroute on the servers


```python
for node_name in node_names:
    print(node_name)
    node = slice.get_node(name=node_name)
    stdout, stderr = node.execute(f'sudo apt install traceroute', quiet=True)  
```

## IP Route Population

```python
for node_idx, (server_name, switch_name) in enumerate(zip(server_names, switch_names)):
    server = slice.get_node(name=server_name)
    switch = slice.get_node(name=switch_name)
    print(switch_name)
        
    #Fill static routes in the IP table       
    for subnet_entry in switch_route_conf[node_idx+1].keys():
        gateway_entry = switch_route_conf[node_idx+1][subnet_entry][0]
        device_entry  = switch_route_conf[node_idx+1][subnet_entry][1]
        print("Switch - Subnet entry: %s, Gateway entry: %s, Device entry:%s" % (subnet_entry, gateway_entry, device_entry))
        stdout, stderr = switch.execute(f'sudo ip route del {subnet_entry}', quiet=False)
        #stdout, stderr = switch.execute(f'sudo ip route del {gateway_entry}', quiet=False)
        #if subnet_entry == "default":
        #    print('sudo ip route add %s dev %s'%(subnet_entry, device_entry))
        #    stdout, stderr   = switch.execute(f'sudo ip route add {gateway_entry} dev {device_entry}') 
        #elif subnet_entry != "default":
        print('sudo ip route add %s via %s dev %s'%(subnet_entry, gateway_entry, device_entry))
        stdout, stderr   = switch.execute(f'sudo ip route add {subnet_entry} via {gateway_entry} dev {device_entry} onlink')       
    
    #Fill static routes in the IP table
    server_ip_route_subnet  = "default"
    server_ip_route_gateway = server_route_conf[node_idx+1]["default"][0]
    server_ip_route_dev = server_route_conf[node_idx+1]["default"][1]
    print("Server - Subnet entry: %s, Gateway entry: %s, Device entry:%s" % (server_ip_route_subnet, server_ip_route_gateway, server_ip_route_dev))
    stdout, stderr = server.execute(f'sudo ip route del {server_ip_route_subnet}', quiet=False)
    ###stdout, stderr   = server.execute(f'sudo ip route {server_ip_route_gateway} dev {server_ip_route_dev}') #NOT FOR SERVERS
    print('sudo ip route add %s via %s dev %s'%(server_ip_route_subnet, server_ip_route_gateway, server_ip_route_dev))
    stdout, stderr   = server.execute(f'sudo ip route add {server_ip_route_subnet} via {server_ip_route_gateway} dev {server_ip_route_dev} onlink')
    
    #Enable forwarding on the switches
    stdout, stderr = switch.execute(f'sudo sysctl -w net.ipv4.ip_forward=1', quiet=True)
    stdout, stderr = server.execute(f'sudo sysctl -w net.ipv4.ip_forward=1', quiet=True)
```

### Draw the network topology

The following cell will draw the network topology, for your reference. The interface name and addresses of each experiment interface will be shown on the drawing.

```python
l2_nets = [(n.get_name(), {'color': 'lavender'}) for n in slice.get_l2networks() ]
print(l2_nets)
l3_nets = [(n.get_name(), {'color': 'pink'}) for n in slice.get_l3networks() ]
print(l3_nets)
hosts   =   [(n.get_name(), {'color': 'lightblue'}) for n in slice.get_nodes()]
print(hosts)
nodes = l2_nets + l3_nets + hosts
ifaces = [iface.toDict() for iface in slice.get_interfaces()]
edges = [(iface['network'], iface['node'], 
          {'label': iface['physical_dev'] + '\n' + iface['ip_addr'] + '\n' + iface['mac']}) for iface in ifaces]
```

```python
import networkx as nx
import matplotlib.pyplot as plt
plt.figure(figsize=(len(nodes),len(nodes)))
G = nx.Graph()
G.add_nodes_from(nodes)
G.add_edges_from(edges)
pos = nx.spring_layout(G, k = 0.1)
nx.draw(G, pos, node_shape='s',  
        node_color=[n[1]['color'] for n in nodes], 
        node_size=[len(n[0])*400 for n in nodes],  
        with_labels=True);
nx.draw_networkx_edge_labels(G,pos,
                             edge_labels=nx.get_edge_attributes(G,'label'),
                             font_color='gray',  font_size=12, rotate=False);

plt.savefig("network_topo_graph.png", format="PNG")
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create an empty undirected graph
G = nx.Graph()

# Add nodes with labels
node_labels = {
    1: "1 LOSA",
    2: "2 SALT",
    3: "3 STAR",
    4: "4 ATLA",
    5: "5 DALL",
    6: "6 KANS",
}
G.add_nodes_from(node_labels.keys())

# Define the edges
edges = [(1, 2), (1, 5), (2, 3), (2, 6), (3, 6), (3, 4), (4, 5), (5, 6)]
G.add_edges_from(edges)

# Draw the graph with labels
pos = nx.spring_layout(G, k=0.3)  # Position the nodes using a layout algorithm
nx.draw(G, pos, with_labels=False, node_size=500, node_color="lightblue")
nx.draw_networkx_labels(G, pos, labels=node_labels, font_size=10, font_color="black")

# Display the graph
plt.axis("off")
plt.savefig("networkx_graph.png", format="PNG")
plt.show()
```
